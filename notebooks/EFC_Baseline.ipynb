```python
# EFC_Baseline.ipynb

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from src.efc_core import efc_potential, expansion_rate

# 1️⃣ Parameter space
rho = np.linspace(1e-27, 1e-24, 100)
S = np.linspace(0.1, 0.9, 100)

# 2️⃣ Compute energy flow and expansion
Ef = efc_potential(rho, S)
H = expansion_rate(Ef, S)

# 3️⃣ Prepare results for export
results = pd.DataFrame({
    "Entropy_S": S,
    "Density_rho": rho,
    "EnergyFlow_Ef": Ef,
    "ExpansionRate_H": H
})

# 4️⃣ Save numeric results
results.to_csv("../output/EFC_Baseline_results.csv", index=False)

# 5️⃣ Visualize
plt.figure(figsize=(8, 5))
plt.plot(S, H, color='blue', label='EFC Expansion Rate (H)')
plt.xlabel('Entropy (S)')
plt.ylabel('Expansion Rate (H, arbitrary units)')
plt.title('EFC Baseline: Expansion vs. Entropy')
plt.legend()
plt.grid(True)
plt.tight_layout()

# 6️⃣ Save plot
plt.savefig("../output/EFC_Baseline_plot.png", dpi=300, bbox_inches='tight')
plt.show()

# 7️⃣ Summary stats
H_LCDM = np.mean(H) * 0.7
print(f"EFC mean H: {np.mean(H):.3e}")
print(f"ΛCDM (normalized) H: {H_LCDM:.3e}")
print("Output saved to: /output/EFC_Baseline_results.csv and /output/EFC_Baseline_plot.png")
```
