```python
import numpy as np
import matplotlib.pyplot as plt
from src.efc_core import efc_potential, expansion_rate

rho = np.linspace(1e-27, 1e-24, 100)
S = np.linspace(0.1, 0.9, 100)
Ef = efc_potential(rho, S)
H = expansion_rate(Ef, S)

plt.figure(figsize=(8,5))
plt.plot(S, H, color='blue', label='EFC Expansion Rate (H)')
plt.xlabel('Entropy (S)')
plt.ylabel('Expansion Rate (H, arbitrary units)')
plt.title('EFC Baseline: Expansion vs. Entropy')
plt.legend()
plt.grid(True)
plt.show()

H_LCDM = np.mean(H) * 0.7
print(f"EFC mean H: {np.mean(H):.3e}")
print(f"ΛCDM (normalized) H: {H_LCDM:.3e}")
```
